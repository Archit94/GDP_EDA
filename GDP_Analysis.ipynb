## Part I-A


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("./Statewise GDP.csv")
# finding the percentage null values in each row
row_null_perc = (df.isnull().sum(axis=1)/len(df.columns))*100
row_null_perc
```

<b>Remove the rows: '(% Growth over the previous year)' and 'GSDP - CURRENT PRICES (` in Crore)' for the year 2016-17</b>

```python
# Since the % of missing values is very high in 2016-17 rows, we will drop them
df = df[row_null_perc < 60]
df
```

```python
# finding columnwise null %
col_null_perc = (df.isnull().sum(axis=0)/len(df.index))*100
col_null_perc
```

```python
# as we can see that for West Bengal, all are null values, we can drop that column
df=df.drop(['West Bengal1'],axis=1)
df
```

<b>Calculate the average growth of states for the duration 2013-14, 2014-15 and 2015-16</b>

```python
# we need to do analysis on only % Growth part so we can filter out the rows having 'GSDP - Current Prices'
df1=df[df['Items  Description'] == '(% Growth over previous year)']

# Dropping the Items Description column as it will have has same values
# Also dropping 2012-13 row
df1=df1.drop(['Items  Description'],axis=1)
df1=df1[df1.Duration != '2012-13']

# defining new row with avg values of % growth
#since some values are NaN, but we can't fill them with the avg of other states' GSDP or median, thus retaining them as it is 
df1.loc['avg_growth'] = df1.mean()
df1.loc['avg_growth','Duration']='avg_growth'
df1
```

```python
# Transposing data and making first row as col headers
df_t=df1.T
head=df_t.iloc[0,:]
df_t=df_t.iloc[1:]
df_t.columns=head
df_t
```

```python
#sorting dataframe on basis of avg_growth 
df_t=df_t.sort_values(['avg_growth'])
df_t
```

<b> Plot avg growth for the states</b>

```python
# plotting graph
# since there are many states we can take them on y-axis
plt.figure(figsize=(12,12))
bp=sns.barplot(df_t.avg_growth,df_t.index)
bp.set_title("Avg Growth of States' GSDP")
plt.savefig('avg_growth.png')
plt.show()
```

```python
# We can see from above that north eastern states have fairly high rate of growth, except sikkim and meghalaya which alongwith
# odisha, goa are relatively growing very slow

# My home state UP is above avg. of All India GDP
```

<b>Plot the total GDP of the states for the year 2015-16</b>

```python
# Getting total GSDP for year 2015-16
df2=df[(df['Items  Description'] != '(% Growth over previous year)') & (df['Duration'] == '2015-16')]

# transposing data to get data for graph
df2=df2.T

# modifying column names
header=df2.iloc[0,]
header.iloc[0]='GSDP(crore)'
df2=df2.iloc[2:,]
df2.columns=header

# sorting on basis of GSDP and dropping NA values
df2=df2.sort_values(df2.columns[0])
df2.dropna(inplace=True)

# plotting graph for GSDP of states in log scale (as values vary on a large scale)
plt.figure(figsize=(12,12))
bp2=sns.barplot(df2['GSDP(crore)'],df2.index)
bp2.set_title("GSDP of States")
bp2.set_xscale('log')
plt.savefig('GSDP.png')
plt.show()
```

```python
# We can see that 5 highest GDP states are as follows:
# 1.Tamil Nadu
# 2.UP
# 3.Karnataka
# 4.Gujarat
# 5.Andhra Pradesh

# Bottom-5 consist 2 Union Territories
# 1.Sikkim
# 2.Arunachal Pradesh
# 3.Puducherry
# 4.Meghalaya
# 5.Chandigarh
```

# Part I-B

```python
# Reading files list using glob
import glob
filenames = glob.glob("./NAD*.csv")
l=[]
print(filenames)
```

<b>Reading many dataframes and combining</b>

```python
# combining all files in a single dataframe
for i in filenames:
    data = pd.read_csv(i,encoding='latin_1')
    # extracting only Item and 2014-15 columns
    data = data[['Item','2014-15']]
    
    # since some data  has * at the end, removing same for resolving conflict during merging
    data['Item']=data['Item'].map(lambda x: x.rstrip('*'))
    # taking transposed data and making first row as header in new dataframe
    data = data.T
    head = data.iloc[0]
    data = data.iloc[1:]
    data.columns = head
    
    # adding state column and making it the first column
    data['State'] = i.split('-')[1]
    col = data.columns.tolist()[-1:] + data.columns.tolist()[:-1]
    data = data[col]
    l.append(data)

# final dataframe has all the previous dataframes combined
final_df = pd.concat(l, axis=0,sort=False,ignore_index=True)
final_df
```

<b>Removing Union Territories</b>

```python
# removing the union territories
UT=['Andaman_Nicobar_Islands','Puducherry','Chandigarh','Delhi']
final_df=final_df[~final_df.State.isin(UT)]
final_df.State
```

<b>Plotting the GDP per capita for all the states</b>

```python
# plotting per capita gsdp
final_df1=final_df.sort_values(['Per Capita GSDP (Rs.)'] )
plt.figure(figsize=(12,12))
bar_p=sns.barplot(final_df1['Per Capita GSDP (Rs.)'],final_df1.State)
bar_p.set_title("Per capita GDP of States")
plt.savefig('Per_capita_GDP.png')
plt.show()
```

```python
# From above Graph,
# Top 5 states: Goa, Sikkim, Haryana, Kerala, Uttarakhand
# Reasons can be attributed to lower population but high gsdp
# Bottom 5 states: Bihar, Uttar Pradesh, Manipur, Assam, Jharkhand
# For these, one main reason is high population which is unskilled and other is lack of production industries
```

```python
# finding ratio of max/min per capita GDP
rat = final_df1['Per Capita GSDP (Rs.)'].max()/final_df1['Per Capita GSDP (Rs.)'].min()
print(rat)
```

```python
# The ratio of highest and lowest per capita GDP of states is very high indicating the irregular distribution
# of talent, money and administration in different parts of India
```

<b>Plotting the percentage contribution of the primary, secondary and tertiary sectors as a percentage of the total GDP</b>

```python
# Making new columns for % contribution of sectors
final_df1['Primary_perc']=final_df1['Primary']/final_df1['TOTAL GSVA at basic prices'] *100
final_df1['Secondary_perc']=final_df1['Secondary']/final_df1['TOTAL GSVA at basic prices'] *100
final_df1['Tertiary_perc']=final_df1['Tertiary']/final_df1['TOTAL GSVA at basic prices'] *100
final_df1['Misc_perc']=100-(final_df1['Tertiary_perc']+final_df1['Secondary_perc']+final_df1['Primary_perc'])

# plot
barWidth = 0.85
plt.figure(figsize=(40,20))
# Draw Primary Bars
plt.bar(np.arange(len(final_df1.index)), final_df1.Primary_perc, color='#b5ffb9', edgecolor='white', width=barWidth,label="Primary %")
# Draw Secondary Bars
plt.bar(np.arange(len(final_df1.index)), final_df1.Secondary_perc, bottom=final_df1.Primary_perc, color='#f9bc86', edgecolor='white'\
        , width=barWidth,label="Secondary %")
# Draw Tertiary Bars
plt.bar(np.arange(len(final_df1.index)), final_df1.Tertiary_perc, bottom=[i+j for i,j in zip(final_df1.Secondary_perc,\
            final_df1.Primary_perc)], color='#a3acff', edgecolor='white', width=barWidth,label="Tertiary %")
 
# Change the name of x axis
plt.xticks(np.arange(len(final_df1.index)),final_df1.State)
plt.xlabel("% wise contribution of sectors")
plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)
# Show the graph
plt.savefig('perc_sector.png')
plt.show()
```

```python
# We can see that for high per capita GDP states, majority of sector is Secondary and Tertiary
# Primary consists of the least percentage.
# Whereas for less GSDP per capita states, agriculture comes out to be 1st or 2nd high contributing sector, 
# whereas Secondary sector contribution is very less
```

<b>Categorise the states into four groups based on the GDP per capita</b>

```python
# getting quantiles for specified quantile values
q1,q2,q3,q4=final_df1['Per Capita GSDP (Rs.)'].quantile([0.20, 0.5, 0.85, 1])
print(q1,q2,q3,q4)
```

```python
# function to categorize states in C1..C4

def categorize(x):
    if(x<=q1):
        return 'C4'
    if(x<=q2 and x>q1):
        return 'C3'
    if(x<=q3 and x>q2):
        return 'C2'
    if(x>q3):
        return 'C1'

#applying categorize function to create a category column

final_df1['Category'] = final_df1['Per Capita GSDP (Rs.)'].apply(categorize)
```

```python
final_df1
```

<b>Find the top 3/4/5 sub-sectors</b>

```python
# Find the contribution of sub sectors
sub_sec = ['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services',\
           'Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting'\
           ,'Financial services','Real estate, ownership of dwelling & professional services','Public administration','Other services','TOTAL GSVA at basic prices']

final = final_df1.groupby('Category')[sub_sec].sum()
# Calculating % contribution of sub-sectors by dividing by total gsva
for i in sub_sec:
    final[i]=final[i]/final[sub_sec[-1]] *100
cat= final.index

# Transposing for plotting for each category
final=final.T
```

<b>Plotting the contribution of the sub-sectors as a percentage of the GSDP of each category. </b>

```python
# Here, we are finding top categories that contribute to approx. 80% of the GSDP of respective categories of states
# Also, plotting sub-sector wise contribution for each category
li=[]

for i in final.columns:
    sorted_ = pd.DataFrame(final.loc[:,i].sort_values(ascending=False).iloc[1:,])
    # finding the top 80 % contributors
    sorted_['Cum_sum'] = np.cumsum(sorted_,axis=0)
    print(sorted_.Cum_sum)
    res = list(filter(lambda i: i > 80.0, (sorted_.Cum_sum).tolist()))[0]
    li.append(sorted_.index[:(sorted_.Cum_sum).tolist().index(res)+1])
    
    # Plotting the figure
    plt.figure(figsize=(8,6))
    plt.title(i + ' - Sub-Sector wise contribution')
    y_pos = np.arange(len(sorted_.index))    
    plt.barh(y_pos, sorted_[i])
    plt.yticks(y_pos,sorted_.index)
    plt.savefig('perc_sub_' + str(i) + '.png')
    plt.show()
```

```python
# printing the top 80% contributors for GSDP category wise
for i in range(4):
    print('TOP CONTRIBUTORS FOR C'+str(i+1))
    print(list(li[i]))
    print()
```

```python
# It can be easily seen that C1,C2 states have mix % of Manufacturing, Agriculture, Trade, Real State and Construction as top
# contributors.
# This is lacking in C3 and C4 states
```

<b>Preparing and Manipulating data for easy merging</b>

```python
# Reading the drop_out data
drop_out= pd.read_csv('rs_session243_au570_1.1.csv')
drop_out = drop_out[['Level of Education - State','Primary - 2014-2015.1', 'Upper Primary - 2014-2015','Secondary - 2014-2015']]

# Renaming to State and removing .1 from Primary - 2014-2015 column name
drop_out.columns=['State','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']
drop_out
```

```python
# Changing the relevant differences in State names(not touching Union Territories, as they are not present in I-B data)
drop_out.loc[34,'State']='Uttarakhand'
drop_out.loc[6,'State']='Chhattisgarh'
final_df1['State']=final_df1['State'].str.replace('_',' ')
final_df1.loc[12,'State']='Jammu and Kashmir'

# Merging the Dataframes on State
drop_out_final=pd.merge(final_df1, drop_out, how='left', on=['State'])
drop_out_final
```

<b>Correlation of GDP per capita with dropout rates in education</b>

```python
#Drawing Scatter plots with regression line for observing the correlation of 'Per Capita GSDP (Rs.)' with dropout rates
sns.lmplot(x='Primary - 2014-2015',y='Per Capita GSDP (Rs.)',data=drop_out_final)
k=drop_out_final['Per Capita GSDP (Rs.)'].astype(np.float64)
print('correlation with Primary dropout ',drop_out_final['Primary - 2014-2015'].corr(k))
plt.savefig('corr_primary.png')
plt.show()
```

```python
sns.lmplot(x='Upper Primary - 2014-2015',y='Per Capita GSDP (Rs.)',data=drop_out_final)
print('correlation with Upper Primary dropout ',drop_out_final['Upper Primary - 2014-2015'].corr(k))
plt.savefig('corr_upper_primary.png')
plt.show()
```

```python
sns.lmplot(x='Secondary - 2014-2015',y='Per Capita GSDP (Rs.)',data=drop_out_final)
print('correlation with Secondary dropout ',drop_out_final['Secondary - 2014-2015'].corr(k))
plt.savefig('corr_secondary.png')
plt.show()
```

```python
# We can observe that dropout rates in each category are inversely correlated to per capita GSDP.
# Thus we can conclude that education is a key area to focus for preparing skilled population that can contribute to state GDP.
```
